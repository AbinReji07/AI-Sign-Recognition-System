```python
import pandas as pd
```

```python
df=pd.read_csv("/content/lung_cancer_examples.csv")
df
```

```python
df.drop(["Name","Surname"],inplace=True,axis=1)
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]#y=df["Result"]
y
```

```python
#scaling in machine learning

#minmaxscaler,standard scaler

#sklearn--sci kit learn
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()#object creation
scaler.fit(X)#
```

```python
X_scaled=scaler.transform(X)
X_scaled
```

```python
#spliting for training and testing

#training= 70
#testing = 30
```

```python
#X,y
#X_scaled.y-output
#79/100---0.7---training
#30/100---0.3---testing

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
#by setting a erandom state ,you ensure that the sequence of random numbers is the same each time you run the code
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)#mathematical model
```

```python
#new value prediction
y_new=knn.predict(scaler.transform([[27,20,2,5]]))
if y_new==1:
  print("suffer cancer")
else:
  print("no cancer")
```

```python
#performance messures
#accuracy score

#X_train          y_train
#X_test           y_test

#knn--- mathematical model
#X_test ,knn
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
y_test.values#original dataset
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
